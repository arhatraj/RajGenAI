```python
import requests
import pdfplumber
from PyPDF2 import PdfReader
```

```python
url = input("Enter PDF link: ")

r = requests.get(url, allow_redirects=True)
```

```python
open('ciena.pdf', 'wb').write(r.content)

pdf = pdfplumber.open("ciena.pdf")

#pdf = pdfplumber.open("Enter path of the file")

# total number of pages
Npages = len(pdf.pages)

print("Total pages: ", Npages)

# Extracting text of each page
for number, pageText in enumerate(pdf.pages):
    print("Page Number: ", number+1)
    print(pageText.extract_text())
```

```python
#Extract images
reader = PdfReader("ciena.pdf")
for page in reader.pages:
    for image in page.images:
        with open(image.name, "wb") as fp:
            fp.write(image.data)
```
