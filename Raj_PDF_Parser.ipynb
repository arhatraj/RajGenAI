```python
pdf = pdfplumber.open("C:/Users/RajPrashantKamble/Downloads/Snowflake_Introduction.pdf")

# total number of pages
Npages = len(pdf.pages)

print("Total pages: ", Npages)

# Extracting text of each page
for number, pageText in enumerate(pdf.pages):
    print("Page Number: ", number+1)
    print(pageText.extract_text())
```

```python
# import beatiful soup for web scrapping

import requests
from bs4 import BeautifulSoup

urls = 'https://www.ciena.com/__data/assets/pdf_file/0025/25369/Waveserver-Ai-DS.pdf'
grab = requests.get(urls)
soup = BeautifulSoup(grab.text, 'html.parser')

# opening a file in write mode
f = open("test1.txt", "w")

# traverse paragraphs from soup
for link in soup.find_all("a"):
    data = link.get('href')
    f.write(data)
    f.write("\n")

f.close()
```

```python
# To Extract images from pdf

from PyPDF2 import PdfReader

reader = PdfReader("C:/Users/RajPrashantKamble/Downloads/Snowflake_Introduction.pdf")
for page in reader.pages:
    for image in page.images:
        with open(image.name, "wb") as fp:
            fp.write(image.data)
```
